```python
#inputs
#[[n1, n2, space1, space2, l1, l2, h1, w1]]
# n1 = 1~9
# n2 = 1~9
# space1 = 
# 
# core loss parameter
core_temp = 70

core_a = 0.6942
core_x = 1.4472
core_y = 2.4769
core_b = 4.7948
core_c = 0.0684
core_d = 4e-4
core_LT = core_b - core_c*core_temp + core_d*core_temp**2

csv_data = pandas.read_csv("core_parameter.csv")

l1_list = csv_data["l1"]
l2_list = csv_data["l2"]
h1_list = csv_data["h1"]
w1_list = csv_data["w1"]


class MyProblem(Problem):
    def __init__(self):
        super().__init__(n_var=7,     #number of inputs
                         n_obj=2,     #number of outputs
                         n_constr=4,  #nubmer of constraints
                         # N1 N2 model d1 d2 M1 M2
                         xl=np.array([8, 8, 0, 0, 0, 1, 1]), #input lower bounds
                         xu=np.array([50, 50, 9, 2, 2, 2, 2])) #input upper bounds


# ["N1","N2","d1","d2","freq","move_tx","move_rx","offset_tx","offset_rx","V1","per","space1","space2","space3","space4","l1", "l2", "h1", "w1", "I1", "I2"]
    def _evaluate(self, X, out, *args, **kwargs):

        I1 = 4.36
        I2 = 3.17

        N1 = X[:,0]
        N2 = X[:,1]
        index = X[:,2]
        index_d1 = X[:,3]
        index_d2 = X[:,4]
        M1 = X[:,5]
        M2 = X[:,6]
        M = M1+M2

        l1 = np.take(l1_list, index, index.astype(np.int64)) * 1e-3
        l2 = np.take(l2_list, index, index.astype(np.int64)) * 1e-3
        h1 = np.take(h1_list, index, index.astype(np.int64)) * 1e-3
        w1 = np.take(w1_list, index, index.astype(np.int64)) * 1e-3

        # 20AWG
        d1_list = [0.812e-3, 1.02e-3, 1.29e-3] # m
        d2_list = [0.812e-3, 1.02e-3, 1.29e-3] # m
        d1A_list = [0.0333, 0.021, 0.0132] # ohm/m
        d2A_list = [0.0333, 0.021, 0.0132] # ohm/m
        dd1_list = [0.23e-3, 0.28e-3, 0.28e-3]
        dd2_list = [0.23e-3, 0.28e-3, 0.28e-3]

        overall_d1 = [1.473e-3, 1.73e-3, 2.06e-3]
        overall_d2 = [1.473e-3, 1.73e-3, 2.06e-3]



        d1 = np.take(d1_list, index_d1.astype(np.int64))
        d2 = np.take(d2_list, index_d2.astype(np.int64))
        d1A = np.take(d1A_list, index_d1.astype(np.int64))
        d2A = np.take(d2A_list, index_d2.astype(np.int64))
        dd1_list = np.take(dd1_list, index_d1.astype(np.int64))
        dd2_list = np.take(dd2_list, index_d2.astype(np.int64))
        overall_d1 = np.take(overall_d1, index_d1.astype(np.int64))
        overall_d2 = np.take(overall_d2, index_d2.astype(np.int64))


        freq = 40e+3
        per = 3000 * 4 * 3.14 * 1e-7
        skin_depth = (1.678/3.14/freq/0.999991/(4*3.14*1e-7))**0.5*1e-3
        wire_length = 2*( (l1+l2) + (w1+l2) )
        DCR_N1 = N1 * wire_length * d1A
        DCR_N2 = N2 * wire_length * d2A

        FR1 = 1/3 * (d1/skin_depth) * (2*M**2 + 1)
        FR2 = 1/3 * (d2/skin_depth) * (2*M**2 + 1)


        # copperloss
        copperloss_N1 = I1**2 * DCR_N1 * FR1
        copperloss_N2 = I2**2 * DCR_N2 * FR2
        copperloss_total = copperloss_N1 + copperloss_N2

        V1 = 459.2
        Req = (5*l1+2*l2+2*h1)/(2*l1*w1)/per
        Lmt = N1**2/Req
        Lmr = N2**2/Req
        mag_current = V1/(2*3.141592*freq)/Lmt
        flux = N1 * mag_current/Req
        B = flux / (2*w1*l1)
        V_core = ((h1 + 2*l1)*(4*l1+2*l2)*(w1) - 2*l2*h1*w1) # Core volue [unit : m^3]
        coreloss = core_a * freq**core_x * B**core_y * core_LT * V_core # coreloss [unit : W]

        total_loss = copperloss_total + coreloss
        total_volume = (h1 + 2*l1)*(4*l1+2*l2)*(w1+l2)

        gh1N1 = (h1 - N1/M1*overall_d1 - 3e-3)
        gh1N2 = (h1 - N2/M2*overall_d2 - 3e-3)
        gB = ((B > 0)&(B < 0.3)).astype(np.float64)-0.5
        ratio = 700/510
        gN2 = -(N2/N1 - (ratio-0.1))*(N2/N1 - (ratio+0.1))

        # loss condition


        out["F"] = np.column_stack([total_volume,total_loss]) # "Minimize" values (volume, coreloss)
        out["G"] = np.column_stack([gh1N1,gh1N2,gB,gN2]) # > 0 Constraints
        out["G"] = - out["G"] # Actually < 0 

problem = MyProblem()
```

```python
from pymoo.algorithms.nsga2 import NSGA2
from pymoo.factory import get_sampling, get_crossover, get_mutation
from pymoo.operators.mixed_variable_operator import MixedVariableSampling, MixedVariableMutation, MixedVariableCrossover

# N1 N2 d1 d2 freq move_tx move_rx, offset_tx, offset_rx , per, space1, space2, space3, spae4, l1, l2, h1, w1
mask = ["int", "int", "int", "int", "int", "int", "int"]
#mask = ["int","int","int","int","int","int","int","int","int","int","int","int","int","int","int","int","int","int"]
#mask = ["real","real","real","real","real","real","real","real","real","real","real","real","real","real","real","real","real","real"]

algorithm = NSGA2(
    pop_size=300,
    n_offsprings=10, # 10

    #sampling=get_sampling("real_random"),
    sampling = MixedVariableSampling(mask, {
        "real": get_sampling("real_random"),
        "int": get_sampling("int_random")
    }),

    #crossover=get_crossover("real_sbx", prob=0.9, eta=15), #Simulated Binary Crossover
    crossover = MixedVariableCrossover(mask, {
        "real": get_crossover("real_sbx", prob=0.7, eta=15),
        "int": get_crossover("int_sbx", prob=0.5, eta=15)
    }),

    #mutation=get_mutation("real_pm", eta=40),
    mutation = MixedVariableMutation(mask, {
        "real": get_mutation("real_pm", eta=3.0),
        "int": get_mutation("int_pm", eta=1)
    }),

    eliminate_duplicates=True
)
```

```python
from pymoo.factory import get_termination

termination = get_termination("n_gen", 30)
```

```python
from pymoo.optimize import minimize

res = minimize(problem,
               algorithm,
               termination,
               seed=1, #RANDOM SEED
               save_history=False,
               verbose=True)
res

#print("Best solution found: %s" % res.X)
#print("Function value: %s" % res.F)
#print("Constraint violation: %s" % res.CV)
```
